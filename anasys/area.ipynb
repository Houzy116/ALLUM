```python
#Area statistics
for landcover_type in ['LCCS1','LCCS2','LCCS3','IGBP']:
    print(landcover_type)
    if landcover_type=='LCCS2':
        lc_len=10
    elif landcover_type=='LCCS3':
        lc_len=9
    elif landcover_type=='IGBP':
        lc_len=16
    elif landcover_type=='LCCS1':
        lc_len=15 
    stat=torch.load(f'/data2/hzy/albedo2/data/stat_sw-mean_sza-85_band-shortwave_{landcover_type}.pth')
    _,sza_L=torch.load(root_path+'sza_and_szaL.pth')
    albedo_all=np.zeros((12,20,lc_len+1,180,360)).astype(np.float32)
    area_all=np.zeros((12,20,lc_len+1,180,360)).astype(np.float32)
    mask=np.zeros((12,20,lc_len+1,180,360))
    for y in range(2001,2021):
        for m in range(1,13):
            albedo_all[m-1,y-2001]=stat[f'{y}-{m}-albedo']
            area_all[m-1,y-2001]=stat[f'{y}-{m}-typearea']
            mask[m-1,y-2001]=np.expand_dims(sza_L[y-2001,m-1],0).repeat(lc_len+1,axis=0)
    albedo_all[albedo_all==0]=np.nan
    albedo_all[mask>=2]=np.nan
    area_all=month_fill(area_all,mask>=2)
    albedo_all=month_fill(albedo_all,mask>=2)
    area_m=np.nansum(area_all,axis=3)
    area_m=np.nansum(area_m,axis=3)
    area=np.nanmean(area_m,axis=0)
    data_m=torch.load(f"/data2/hzy/albedo2/confuse_matrix/confuse_matrix_2020_{landcover_type}.pth")
    mask_t=np.expand_dims(mask[:,0,:],0).repeat(lc_len+1,axis=0).transpose((1,3,4,0,2))
    data_m[mask_t>=2]=np.nan
    data_m=month_fill(data_m,mask_t>=2)
    data_y=np.nanmean(data_m,axis=0)
    transf=np.nansum(data_y,axis=0)
    transf=np.nansum(transf,axis=0)
    nochange=[transf[i,i]/10000 for i in range(lc_len+1)]
    change_sub=[(np.sum(transf,axis=1)[i]-transf[i,i])/10000 for i in range(lc_len+1)]
    change_add=[(np.sum(transf,axis=0)[i]-transf[i,i])/10000 for i in range(lc_len+1)]
    change_add2=[]
    change_sub2=[]
    nochange2=[]
    label=[]
    print('type       add     sub     non-conversion')
    for i in range(lc_len+1):
        if i==1:
            continue
        change_add2.append(change_add[i])
        change_sub2.append(change_sub[i])
        nochange2.append(nochange[i])
        label.append(type_code[landcover_type][i])
        print(type_code[landcover_type][i].ljust(50,' '),str(change_sub[i]).ljust(20,' '),str(nochange[i]).ljust(20,' '),str(change_add[i]).ljust(20,' '))
        
    kk=[nochange2[i]+change_add2[i] for i in range(lc_len)]
    print('e4 km2')
    print('total areas:',(np.array(nochange2).sum()+np.array(change_add2).sum()))
    print('conversion areas:',np.array(change_add2).sum())
    print('non-conversion areas:',np.array(nochange2).sum())
    print('conversion fraction:',np.array(change_add2).sum()/(np.array(nochange2).sum()+np.array(change_add2).sum())*100,'%')
```

```python
for landcover_type in ['LCCS1','LCCS2','LCCS3','IGBP']:
    print(landcover_type)
    if landcover_type=='LCCS2':
        lc_len=10
    elif landcover_type=='LCCS3':
        lc_len=9
    elif landcover_type=='IGBP':
        lc_len=16
    elif landcover_type=='LCCS1':
        lc_len=15 
    stat=torch.load(f'/data2/hzy/albedo2/data/stat_sw-mean_sza-85_band-shortwave_{landcover_type}.pth')
    _,sza_L=torch.load(root_path+'sza_and_szaL.pth')
    albedo_all=np.zeros((12,20,lc_len+1,180,360)).astype(np.float32)
    area_all=np.zeros((12,20,lc_len+1,180,360)).astype(np.float32)
    mask=np.zeros((12,20,lc_len+1,180,360))

    for y in range(2001,2021):
        for m in range(1,13):
            albedo_all[m-1,y-2001]=stat[f'{y}-{m}-albedo']
            area_all[m-1,y-2001]=stat[f'{y}-{m}-typearea']
            mask[m-1,y-2001]=np.expand_dims(sza_L[y-2001,m-1],0).repeat(lc_len+1,axis=0)
    albedo_all[albedo_all==0]=np.nan
    albedo_all[mask>=2]=np.nan
    area_all=month_fill(area_all,mask>=2)
    albedo_all=month_fill(albedo_all,mask>=2)
    area_m=np.nansum(area_all,axis=3)
    area_m=np.nansum(area_m,axis=3)
    area=np.nanmean(area_m,axis=0)
    
    data_m=torch.load(f"/data2/hzy/albedo2/confuse_matrix/confuse_matrix_2020_{landcover_type}.pth")
    mask_t=np.expand_dims(mask[:,0,:],0).repeat(lc_len+1,axis=0).transpose((1,3,4,0,2))
    data_m[mask_t>=2]=np.nan
    data_m=month_fill(data_m,mask_t>=2)
    data_y=np.nanmean(data_m,axis=0)
    transf=np.nansum(data_y,axis=0)
    transf=np.nansum(transf,axis=0)
    transf2=transf.copy()
    
    transf22=np.zeros((lc_len+1,lc_len+1))
    for i in range(lc_len+1):
        transf22[i,i]=0
    for x in range(lc_len+1):
        for y in range(lc_len+1):
            transf22[x,y]=(transf2[y,x]-transf2[x,y])
            transf22[y,x]=0
    for i in range(lc_len+1):
        transf22[i,i]=transf2[i,i]
    transf22[transf22==0]=np.nan
    transf3=np.zeros((lc_len,lc_len))
    transf3[:1,:1]=transf22[:1,:1]
    transf3[1:,:1]=transf22[2:,:1]

    transf3[:1,1:]=transf22[:1,2:]
    transf3[1:,1:]=transf22[2:,2:]

    ll=[type_code[landcover_type][i] for i in range(lc_len+1) if i!=1]
    transf3[np.isnan(transf3)]=0
    for i in range(lc_len+1):
        transf[i,i]=0
    transf=np.delete(transf,1,axis=0)
    transf=np.delete(transf,1,axis=1)
    fig, ax = plt.subplots(1, figsize=(20, 20),dpi=600)
    chord_diagram(transf, ll, gap=0.03, use_gradient=False, sort='size', directed=True,
                    cmap='tab20', chord_colors=None, rotate_names=True, fontcolor='grey',ax=ax,fontsize=25)
    # plt.savefig(f'../figure/chord_{landcover_type}.pdf',dpi=600)
    plt.show()
```
